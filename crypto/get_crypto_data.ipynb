```python
import requests

def get_crypto_data_in_inr(coin_id):
    url = f"https://api.coingecko.com/api/v3/coins/{coin_id}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        print(f"Name: {data['name']}")
        print(f"Symbol: {data['symbol']}")
        print(f"Current Price (INR): ₹{data['market_data']['current_price']['inr']}")
        print(f"Market Cap (INR): ₹{data['market_data']['market_cap']['inr']}")
        print(f"24h High (INR): ₹{data['market_data']['high_24h']['inr']}")
        print(f"24h Low (INR): ₹{data['market_data']['low_24h']['inr']}")
    else:
        print(f"Failed to fetch data for coin: {coin_id}")

# Example: Get data for Uniswap (DeFi coin)
get_crypto_data_in_inr("uniswap")

```

```python
import requests

def get_defi_coins_in_inr():
    url = "https://api.coingecko.com/api/v3/coins/markets"
    params = {
        'vs_currency': 'inr',
        'category': 'decentralized-finance-defi',
        'order': 'market_cap_desc',
        'per_page': 10,  # You can increase this up to 250
        'page': 1,
        'sparkline': 'false'
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        coins = response.json()
        for coin in coins:
            print(f"\nName: {coin['name']}")
            print(f"Symbol: {coin['symbol']}")
            print(f"Current Price (INR): ₹{coin['current_price']}")
            print(f"Market Cap (INR): ₹{coin['market_cap']}")
            print(f"24h High (INR): ₹{coin['high_24h']}")
            print(f"24h Low (INR): ₹{coin['low_24h']}")
    else:
        print("Failed to fetch DeFi coin data.")

# Fetch DeFi coins
get_defi_coins_in_inr()
```
